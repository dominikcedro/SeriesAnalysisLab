This task was focused on analysing temperature dataset from St.Louis.

```python
import pandas as pd
import numpy as np
import tabulate

# import data
df = pd.read_csv('data_st_louis.csv', skiprows=4)
```

```python
#find missing values that are coded with -99 and replace them with NaN
df = df.replace(-99, np.nan)
df = df.interpolate()
```

```python
# convert index to datetime
df['Datetime'] = pd.to_datetime(df['Date'], format='%Y%m')
# drop Date column
df = df.drop('Date', axis=1)
df = df.set_index('Datetime')
```

```python
# plot avergae temp series
import matplotlib.pyplot as plt
df['Value'].plot(color='red')
plt.title('Average Temperature in St. Louis')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

```python
#plot histogram of average temp
df['Value'].hist(color='red')
plt.title('Average Temperature in St. Louis')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# plot kernel density plot
df['Value'].plot(kind='kde', color='red')
plt.title('Average Temperature in St. Louis')
plt.xlabel('Temperature')
plt.ylabel('Density')
plt.show()
```

```python
# generate descriptive statistics
print(df['Value'].describe())
```
